### House price pridiction

```python
from sklearn.datasets import load_boston
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
data =load_boston()
```

```python
data
```

```python
dataset = pd.DataFrame(data.data)
```

```python
dataset.columns = data.feature_names
```

```python
dataset
```

```python
dataset.head()
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,:-1] #independent feature
y = dataset.iloc[:,-1] # dependent feature
```

```python
X.head()
```

```python
y.head()
```

## Split train test 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model  import Ridge
from sklearn.model_selection import GridSearchCV
ridge=Ridge()
params ={'alpha':[1e-15,1e-10,1e-3,1,5,10,20,25,30,40,50,60,70,80,100]}
ridge_regression = GridSearchCV(ridge,params,scoring='neg_mean_squared_error',cv=10)
ridge_regression.fit(X_train,y_train)
print(ridge_regression.best_params_)
print(ridge_regression.best_score_)
```

```python
y_pred = ridge_regression.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_pred,y_test)
r2_score
```
